```python

import tkinter as tk
import pandas as pd
import time

class DescriptionApp:
    def __init__(self, master):
        self.master = master
        self.master.title("Description App")
        self.master.configure(background='#272727')  # Set default background color
        self.theme_var = tk.StringVar(value="Dark")  # Default theme is Dark
        self.text_size_var = tk.StringVar(value="Medium")  # Default text size is Medium
        self.text_font_var = tk.StringVar(value="Helvetica")  # Default text font is Helvetica

        # Menu for theme selection
        self.menu_bar = tk.Menu(master)
        self.theme_menu = tk.Menu(self.menu_bar, tearoff=0)
        self.theme_menu.add_radiobutton(label="Dark", variable=self.theme_var, value="Dark", command=self.change_theme)
        self.theme_menu.add_radiobutton(label="Light", variable=self.theme_var, value="Light", command=self.change_theme)
        self.menu_bar.add_cascade(label="Theme", menu=self.theme_menu)

        # Menu for text size selection
        self.text_size_menu = tk.Menu(self.menu_bar, tearoff=0)
        self.text_size_menu.add_radiobutton(label="Small", variable=self.text_size_var, value="Small", command=self.change_text_size)
        self.text_size_menu.add_radiobutton(label="Medium", variable=self.text_size_var, value="Medium", command=self.change_text_size)
        self.text_size_menu.add_radiobutton(label="Large", variable=self.text_size_var, value="Large", command=self.change_text_size)
        self.menu_bar.add_cascade(label="Text Size", menu=self.text_size_menu)

        # Menu for text font selection
        self.text_font_menu = tk.Menu(self.menu_bar, tearoff=0)
        self.text_font_menu.add_radiobutton(label="Helvetica", variable=self.text_font_var, value="Helvetica", command=self.change_text_font)
        self.text_font_menu.add_radiobutton(label="Arial", variable=self.text_font_var, value="Arial", command=self.change_text_font)
        self.text_font_menu.add_radiobutton(label="Times New Roman", variable=self.text_font_var, value="Times New Roman", command=self.change_text_font)
        self.menu_bar.add_cascade(label="Text Font", menu=self.text_font_menu)

        self.master.config(menu=self.menu_bar)

        # Frame for description elements
        self.description_frame = tk.Frame(master, bg='#272727')
        self.description_frame.pack(pady=20, padx=10, side=tk.LEFT)

        self.title_label = tk.Label(self.description_frame, text="Enter Book Title:", font=("Helvetica", 14), bg='#272727', fg='white')
        self.title_label.pack(pady=10)

        self.title_entry = tk.Entry(self.description_frame, font=("Helvetica", 14), bg='#272727', fg='white')  # Apply theme to entry widget
        self.title_entry.pack(pady=5)

        # Frame for dashboard buttons
        self.dashboard_frame = tk.Frame(master, bg='#272727')
        self.dashboard_frame.pack(pady=20, padx=10, side=tk.LEFT)

        self.show_description_button = tk.Button(self.dashboard_frame, text="Show Description", font=("Helvetica", 14),
                                                 command=self.show_description, bg='#4CAF50', fg='white')  # Green button
        self.show_description_button.pack(pady=10)

    def show_description(self):
        book_title = self.title_entry.get()
        filtered_df = self.load_descriptions()
        if filtered_df is None:
            return

        filtered_df = filtered_df[self.descriptions_df['title'] == book_title]
        if len(filtered_df) == 0:
            print(f"No description found for book title: {book_title}")
            return
        
        description = filtered_df.iloc[0]["description"]
        self.description_label = tk.Label(self.description_frame, text="", font=(self.text_font_var.get(), self.get_font_size()), wraplength=400, bg='#272727', fg='white')
        self.description_label.pack()

        self.animate_description(description)

    def load_descriptions(self):
        try:
            self.descriptions_df = pd.read_csv("books_1.Best_Books_Ever.csv")
            return self.descriptions_df
        except FileNotFoundError:
            print("CSV file not found!")
            self.master.destroy()
            return None

    def animate_description(self, description):
        self.description_label.config(text="")
        self.master.update()

        for char in description:
            self.description_label.config(text=self.description_label.cget("text") + char)
            self.master.update()
            time.sleep(0.02)

    def change_theme(self):
        theme = self.theme_var.get()
        if theme == "Dark":
            bg_color = '#272727'
            fg_color = 'white'
        elif theme == "Light":
            bg_color = '#f0f0f0'
            fg_color = 'black'

        # Apply theme to elements
        self.master.configure(background=bg_color)
        self.description_frame.configure(bg=bg_color)
        self.dashboard_frame.configure(bg=bg_color)
        self.title_label.configure(bg=bg_color, fg=fg_color)
        self.title_entry.configure(bg=bg_color, fg=fg_color)
        self.show_description_button.configure(bg='#4CAF50', fg='white')  # Button stays green in both themes
        if hasattr(self, 'description_label'):  # Check if description label exists
            self.description_label.configure(bg=bg_color, fg=fg_color)
        if hasattr(self, 'text_font_var') and hasattr(self, 'text_size_var'):
            font = (self.text_font_var.get(), self.get_font_size())
            if hasattr(self, 'description_label'):
                self.description_label.configure(font=font)

    def change_text_size(self):
        if hasattr(self, 'text_font_var'):
            font = (self.text_font_var.get(), self.get_font_size())
            if hasattr(self, 'description_label'):
                self.description_label.configure(font=font)

    def change_text_font(self):
        if hasattr(self, 'text_size_var'):
            font = (self.text_font_var.get(), self.get_font_size())
            if hasattr(self, 'description_label'):
                self.description_label.configure(font=font)

    def get_font_size(self):
        size = 12  # Default font size
        text_size = self.text_size_var.get()
        if text_size == "Small":
            size = 10
        elif text_size == "Medium":
            size = 12
        elif text_size == "Large":
            size = 16
        return size

```

```python
import tkinter as tk
from tkinter import ttk, messagebox
import pandas as pd

class RatingApp:
    def __init__(self, master):
        self.master = master
        self.master.title("Book Rating App")
        self.master.geometry("400x250")
        self.master.resizable(False, False)
        
        # Load data from CSV
        self.load_data()
        
        # Create widgets
        self.create_widgets()
        
    def load_data(self):
        try:
            self.data = pd.read_csv("books_1.Best_Books_Ever.csv")
        except FileNotFoundError:
            print("CSV file not found!")
            self.data = None
    
    def create_widgets(self):
        if self.data is None:
            self.show_error_message("Error", "Data file not found.")
            return
        
        # Create entry for book title
        self.book_title_label = ttk.Label(self.master, text="Enter Book Title:")
        self.book_title_label.pack(pady=(20, 5))
        
        self.book_title_entry = ttk.Entry(self.master, width=30)
        self.book_title_entry.pack()
        
        # Button to show rating
        self.show_rating_button = ttk.Button(self.master, text="Show Rating", command=self.show_rating)
        self.show_rating_button.pack(pady=10)
        
        # Frame to display rating
        self.rating_frame = ttk.Frame(self.master)
        self.rating_frame.pack(pady=10)
        
    def show_rating(self):
        book_title = self.book_title_entry.get()
        if not book_title:
            self.show_error_message("Error", "Please enter a book title.")
            return
        
        # Filter data for the entered book title
        book_data = self.data[self.data['title'] == book_title]
        
        if book_data.empty:
            self.show_error_message("Error", f"No data found for book title: {book_title}")
            return
        
        # Get rating from the first row
        rating = book_data.iloc[0]['rating']
        
        # Clear previous rating display
        for widget in self.rating_frame.winfo_children():
            widget.destroy()
        
        # Create label to display rating
        rating_label = ttk.Label(self.rating_frame, text=f"Rating for '{book_title}': {rating}")
        rating_label.pack()
        
        # Create a canvas for the rating bar
        self.rating_canvas = tk.Canvas(self.rating_frame, width=200, height=20, bg="white")
        self.rating_canvas.pack(pady=5)
        
        # Define colors for the gradient
        color_start = "#FFD700"  # Gold
        color_end = "#FFA500"    # Orange
        
        # Calculate the length of the filled bar based on the rating
        bar_length = rating / 5 * 200
        
        # Draw the filled bar with gradient color
        for i in range(0, int(bar_length), 2):
            color = self.interpolate_color(color_start, color_end, i / bar_length)
            self.rating_canvas.create_line(i, 0, i, 20, fill=color)
        
    def interpolate_color(self, color_start, color_end, ratio):
        r = int(color_start[1:3], 16) * (1 - ratio) + int(color_end[1:3], 16) * ratio
        g = int(color_start[3:5], 16) * (1 - ratio) + int(color_end[3:5], 16) * ratio
        b = int(color_start[5:], 16) * (1 - ratio) + int(color_end[5:], 16) * ratio
        return f"#{int(r):02X}{int(g):02X}{int(b):02X}"

    def show_error_message(self, title, message):
        messagebox.showerror(title, message)
```

```python
class ImageFrame(tk.Frame):
    def __init__(self, parent, urls):
        super().__init__(parent)
        self.urls = urls
        self.images = []
        self.current_image_index = 0
        
        self.setup_ui()

    def setup_ui(self):
        # Create frame 1 for navigation buttons
        frame1 = tk.Frame(self)
        frame1.pack(side=tk.TOP, pady=10)

        # Create buttons for navigation in frame 1
        prev_button = tk.Button(frame1, text="<<<", width=11,height=1,font="arial 14 bold",bg="black",fg="white",command=self.show_previous_image)
        prev_button.pack(side=tk.LEFT)

        next_button = tk.Button(frame1, text=">>>",width=11,height=1,font="arial 14 bold",bg="black",fg="white", command=self.show_next_image)
        next_button.pack(side=tk.RIGHT)
        
        # Create frame 2 for displaying images
        frame2 = tk.Frame(self)
        frame2.pack()

        # Create label to display image in frame 2
        self.image_label = tk.Label(frame2)
        self.image_label.pack()

        # Load and display the first image
        self.load_image(0)

    def load_image(self, index):
        url = self.urls[index]
        response = requests.get(url)
        img = Image.open(BytesIO(response.content))
        img = img.resize((300, 500))  # Resize image if needed
        photo = ImageTk.PhotoImage(img)
        
        # Save reference to avoid garbage collection
        self.images.append(photo)
        
        # Display image
        self.image_label.config(image=photo)
        self.image_label.image = photo

    def show_previous_image(self):
        if self.current_image_index > 0:
            self.current_image_index -= 1
            self.load_image(self.current_image_index)

    def show_next_image(self):
        if self.current_image_index < len(self.urls) - 1:
            self.current_image_index += 1
            self.load_image(self.current_image_index)
```
